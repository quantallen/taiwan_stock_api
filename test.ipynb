```python
from datetime import datetime, timedelta,date
def datetime_range(start, end, delta):
    current = start
    while current < end :
        yield current
        current += delta
dts  =[dt.striftime('%Y%m-%d T%H:%M:%S') for dt in datetime_range(datetime(datetime.now().year,date.month(),date.day(),9,0,0),datetime(date.year,date.month,date.day,13,25,0),timedelta(minutes = 1))]
print(dts)
```
